```python
filepath = '/Users/tomevans/Downloads/fifa21_validate'
feature_cols = ['total stats', 'base stats', 'movement', 'power', 'skill','mentality', 'attacking', 'defending', 'goalkeeping', 'pac', 'sho', 'pas', 'dri', 'def', 'phy', 'value', 'bp']
target_col = ['ova']

def load_preprocess_data(path):
    df = pd.read_csv(path)
    df.rename(columns=str.lower, inplace=True)
    return df
    
def extract_features_target(df, feature_cols, target_col):
    X = df.loc[:, feature_cols]
    y = df.loc[:, target_col]
    return X, y

def clean_convert_values(X):
    # Define the clean function here
    def clean(x):
        if x in ['CM', 'CAM', 'CDM', 'LM', 'RM']:
            return 'mid'
        elif x in ['RB', 'CB', 'LB', 'LWB', 'RWB']:
            return 'def'
        elif x in ['ST', 'RW', 'LW', 'CF']:
            return 'att'
        else:
            return 'gk'
    X['bp'] = X['bp'].apply(clean)
    label_encoder = LabelEncoder()
    X['bp'] = label_encoder.fit_transform(X['bp'])
    # Your existing code for cleaning 'value' column
    valuefix = X.loc[X['value'].str.contains('\.')]
    valuefix['value'] = valuefix['value'].str.replace('M', '00000')
    X.loc[valuefix.index, 'value'] = valuefix['value']
    X['value'] = (X['value'].str.replace('€', '').str.replace('M', 'e6').str.replace('K', 'e3').str.replace('.', '', regex=False).astype(float))
    return X
    
```

```python
# Define the file path
filepath = '/Users/tomevans/Downloads/fifa21_validate.csv'

# Define the feature columns and target column
feature_cols = ['total stats', 'base stats', 'movement', 'power', 'skill','mentality', 'attacking', 'defending', 'goalkeeping', 'pac', 'sho', 'pas', 'dri', 'def', 'phy', 'value', 'bp']
target_col = 'ova'

# Call the load_preprocess_data function to load and preprocess the data
df = load_preprocess_data(filepath)

# Extract features and target
X, y = extract_features_target(df, feature_cols, target_col)

# Clean and convert values
X_cleaned = clean_convert_values(X)

# Perform further analysis or processing with the cleaned data
X_cleaned
```

```python
predictions_validation = lm.predict(X_cleaned)
print("R2 test score is:",r2_score(y, predictions_validation))
```
